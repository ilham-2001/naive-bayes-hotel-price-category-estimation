```python
import pandas as pd
```

```python
data = pd.read_csv("data/hotel.csv")
data
```

```python
# hilangkan data hotel yang tidak memiliki harga

data = data.drop(data[data.Price == '0'].index)
```

```python
data
```

```python
# hilangkan data hotel yang tidak memiliki rating bintang

data = data.drop(data[data.Class == '-'].index)
```

```python
data
```

```python
for i in data['Facility']:
    data_str = i.strip("[]")
    print(data_str)
    
```

```python
# Punya AC atau tidak?
hasAC = []

for i in data['Facility']:
    data_str = i.strip("[]")

    if "Air conditioning" in data_str:
        hasAC.append(True)
    else:
        hasAC.append(False)

data = data.assign(AirConditioner=hasAC)
data
```

```python
# Punya Bar atau tidak?
hasBar = []

for i in data['Facility']:
    data_str = i.strip("[]")

    if "Bar" in data_str:
        hasBar.append(True)
    else:
        hasBar.append(False)

data = data.assign(Bar=hasBar)
data
```

```python
# Punya Kolam Renang atau tidak?
hasPool = []

for i in data['Facility']:
    data_str = i.strip("[]")

    if "pool" in data_str:
        hasPool.append(True)
    else:
        hasPool.append(False)

data = data.assign(Pool=hasPool)
data
```

```python
# Punya Spa atau tidak?
hasSpa = []

for i in data['Facility']:
    data_str = i.strip("[]")

    if "Spa" in data_str:
        hasSpa.append(True)
    else:
        hasSpa.append(False)

data = data.assign(Spa=hasSpa)
data
```

```python
# Punya Fitnees center atau tidak?
hasFitness = []

for i in data['Facility']:
    data_str = i.strip("[]")

    if "Fitness centre" in data_str:
        hasFitness.append(True)
    else:
        hasFitness.append(False)

data = data.assign(FitenessCenter=hasFitness)
data
```

```python
# Punya Restaurant atau tidak?
hasRestaurant = []

for i in data['Facility']:
    data_str = i.strip("[]")

    if "Restaurant" in data_str:
        hasRestaurant.append(True)
    else:
        hasRestaurant.append(False)

data = data.assign(Restaurant=hasRestaurant)
data
```

```python
# Punya Hot tub atau tidak?
hasHotTub = []

for i in data['Facility']:
    data_str = i.strip("[]")

    if "Hot tub" in data_str:
        hasHotTub.append(True)
    else:
        hasHotTub.append(False)

data = data.assign(HotTub=hasHotTub)
data
```

```python
# ubah price mennjadi integer
price = data['Price']
prices_data = []

for p in price:
    p = p.replace('IDR ', '')
    p = int(p.replace(',', ''))
    prices_data.append(p)

data = data.assign(Price_IDR=prices_data)

data = data.drop(['Price'], axis=1)
data
```

```python
hotelStar = data['Class']
hotelNew = []

for x in hotelStar:
  x = x.replace("-star hotel", "")
  x = int(x)
  hotelNew.append(x)

data = data.assign(hotelClass=hotelNew)
data = data.drop(['Class'], axis=1)

data
```

```python
data = data.drop(['Facility'], axis=1)
data
```

```python
data.to_csv('data/hotel_preprocess_v2.csv')
```
